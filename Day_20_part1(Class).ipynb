```python
# Let's see how to scaled data
# import scaler
from sklearn.preprocessing import StandardScaler
```

```python
# let's create an instance of standardscaler
scaler = StandardScaler()
```

```python
data
```

```python
data_scaled = scaler.fit_transform(data)
```

```python
data_scaled
```

```python
# Now convert 'data_scaled' arry in to dataframe
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
```

```python
# Data Decomposition using PCA
from sklearn.decomposition import PCA
```

```python
# Create an instance of pca with n_compoents =2
pca= PCA(n_components=2)
```

```python
# get pca_df for data (non scaled data)
pca_data = pca.fit_transform(data)
pca_data
```

```python
pca_data = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])
pca_data
```

```python
data.head(2)
```

```python
pca_data.head(2)
```

```python
plt.title("Wine Dataset Before Scaling")
sns.scatterplot(x='PC1', y='PC2', data=pca_data)
```

```python
# 1600  every single dot denotes or repsentes a row i.e a sample of wine given in the dataset
# outlier
```

```python
# Data decomposition for scaled data
pca_scaled = PCA(n_components =2)
```

```python
pca_scaled_data = pca_scaled.fit_transform(data_scaled)
```

```python
pca_scaled_data = pd.DataFrame(pca_scaled_data, columns=['PCA1', 'PCA2'])
```

```python
pca_scaled_data
```

```python
# Side by Side plot the pca of both (non scaled and scaled data)
fig, axes = plt.subplots(1,2, figsize=(12,4))
axes[0].set_title("Wine Dataset Before Scaling")
axes[1].set_title("Wine Dataset After Scaling")
sns.scatterplot(x='PC1', y='PC2', data =pca_data, ax=axes[0])
sns.scatterplot(x='PCA1',  y='PCA2', data=pca_scaled_data, ax=axes[1])
```
