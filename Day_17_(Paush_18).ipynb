```python
data = data[['sex','total_bill', 'tip']]
```

```python
# Sepeta 'X' and 'y'
X = data.drop('sex', axis=1)
```

```python
y= data.sex
```

```python
# import Naivebayes
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
# Fit or train the model
model.fit(X,y)
```

```python
# Make predictions
y_pred = model.predict(X)
```

```python
# Model Accuracy check
from sklearn.metrics import accuracy_score
acs_score = accuracy_score(y, y_pred)
acs_score
```
