```python
# import the library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
data = pd.read_csv(r"/content/drive/MyDrive/Datasets/winequality-red.csv")
```

```python
data.head(2)
```

```python
data.drop('quality', axis=1, inplace=True)
```

```python
data
```

```python
# Scaled the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
data_scaled = scaler.fit_transform(data)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
data_scaled.head()
```

```python
# Decompose the data both(non scaled and scaled)
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=2)
```

```python
pca_data = pca.fit_transform(data)
pca_data = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2'])
pca_data.head()
```

```python
# Decompose the scaled data
data_scaled_pca= pca.fit_transform(data_scaled)
data_scaled_pca = pd.DataFrame(data_scaled_pca, columns=['PCA1', 'PCA2'])
data_scaled_pca.head()
```

```python
# Visual both Non Scaled and Scaled data
fig, axes = plt.subplots(1,2, figsize=(12,5))
axes[0].set_title("2D View of Dataset before scaling")
axes[1].set_title("2D view of Dataset after Scaling")
sns.scatterplot(x='PCA1', y='PCA2', data=pca_data, ax=axes[0])
sns.scatterplot(x='PCA1', y='PCA2', data=data_scaled_pca, ax=axes[1])
```

```python
# Using KMeans Clustering
from sklearn.cluster import KMeans
```

```python
wcss=[]
```

```python
for x in range(2,9):
    k=KMeans(n_clusters=x)
    k.fit_transform(data_scaled_pca)
    wcss.append(k.inertia_)
```

```python
# Plot the wcss values with repsective clusters
plt.figure(figsize=(4,4))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("WCSS Values of Every Possible Clusters")
plt.plot(range(2,9), wcss, marker='D')
```

```python
# From the observation of the above diagram the optimum of cluster is 5
```

```python
kmeans = KMeans(n_clusters = 5 )
```

```python
data_scaled_pca
```

```python
data_scaled['clusters']= kmeans.fit_predict(data_scaled)
```
