```python
# Grid Search
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset tips
data = sns.load_dataset('tips')
```

```python
data.isna().sum()
```

```python
data = data[['total_bill','tip','day']]
```

```python
data.head(2)
```

```python
# Create X and y
X = data.drop('day', axis=1)
y=data['day']
```

```python
X.head(2)
```

```python
# Create train and test data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
# Create a decision tree classifcation model and train it
from sklearn.tree import DecisionTreeClassifier
```

```python
# model
model = DecisionTreeClassifier()
```

```python
# train the model
model.fit(X_train, y_train)
```

```python
# Predict for X_train
y_train_predict = model.predict(X_train)
```

```python
# Predict of X_test
y_test_predict = model.predict(X_test)
```

```python
# Accuracy Score for both of values
from sklearn.metrics import accuracy_score
```

```python
acs_train = accuracy_score(y_train, y_train_predict)
acs_test = accuracy_score(y_test, y_test_predict)
```

```python
acs_train
```

```python
acs_test
```

```python
# Hyperparameter Tuning
```

```python
# Grid Search
# Define Hyperparameter ranges
param_grid = {
    'max_depth': [2, 3, 4, 5, 6, 8, 10],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 4, 8],
    'max_features': [None, 'sqrt', 'log2']
}
```

```python
# Apply GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
```
