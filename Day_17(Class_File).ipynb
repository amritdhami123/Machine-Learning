```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# for this we will use tips dataset
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
# Encode
enc_sex={'Male':1, 'Female':0}
enc_smoker={'Yes':1, 'No':0}
enc_day={'Thur':0, 'Fri':1, 'Sat':2, 'Sun':3  }
enc_time={'Lunch':0, 'Dinner':1}
```

```python
# encode manually
data['sex']=data['sex'].map(enc_sex)
```

```python
data['smoker']= data['smoker'].map(enc_smoker)
```

```python
data['day'] = data['day'].map(enc_day)
```

```python
data['time'] = data['time'].map(enc_time)
```

```python
data.head(10)
```

```python
# Split Data in to 'X' and 'y'
X= data.drop('day', axis=1)
y= data['day']
```

```python
data
```

```python
# Randomly
data.day.unique()
```

```python
data.day.value_counts()
```

```python
# import train test split
from sklearn.model_selection import train_test_split
```

```python
# Create the train and test data set
X_train , X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
# Now Create and fit the model
from sklearn.tree import DecisionTreeClassifier
```

```python
# create model
model = DecisionTreeClassifier()
```

```python
# Predict Values
Data ;_ X_train, X_test
X_train is used for trainit the model
X_train :- Predict
X_test (predict)
```

```python
# Fit the model
model.fit(X_train, y_train)
```

```python
# predict values for X_train
y_train_pred = model.predict(X_train)
```

```python
# Predict valus for X_test
y_test_pred = model.predict(X_test)
```

```python
# Check Accuracy for both
from sklearn.metrics import accuracy_score
```

```python
acs_train = accuracy_score(y_train_pred, y_train)
acs_test = accuracy_score(y_test_pred, y_test)
```

```python
print(f"Accuracy score for Train: {acs_train}")
print(f"Accuracy Score for Test: {acs_test}")
```
